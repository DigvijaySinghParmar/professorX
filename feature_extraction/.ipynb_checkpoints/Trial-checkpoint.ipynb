```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.fft import fft
import pywt
%matplotlib inline
```

```python
# code to generate single frequency signal
def sample_sin(freq):
    sampling_rate = 44100
    duration = 1
    N = sampling_rate*duration
    omega = 2*np.pi*freq
    t = np.linspace(0,N,250)
    y = np.sin(omega*t)
    plt.plot(t,y)
    plt.show
    return y
```

```python
y = sample_sin(16)
```

```python
y_dash = fft(y)
```

```python
plt.plot(abs(y_dash))
```

```python
len(y)
```

```python
x = sample_sin(8)
```

```python
x_dash = fft(x)
```

```python
plt.plot(abs(x_dash))
```

```python
y_new = np.concatenate((y,x), axis = 0)

# y_new is a non-stationary signal
```

```python
y_new_dash = fft(y_new)
```

```python
plt.plot(abs(y_new_dash))
```

```python
plt.plot(y_new)
```

```python
y_dec=pywt.wavedec(y_new, wavelet = 'db4', level=4)
```

```python
plt.plot(y_dec[1])
```

```python
# Seed the random number generator
np.random.seed(0)

time_step = .01
time_vec = np.arange(0, 70, time_step)

# A signal with a small frequency chirp
sig = np.sin(0.5 * np.pi * time_vec * (1 + .1 * time_vec))

plt.figure(figsize=(8, 5))
plt.plot(time_vec, sig)
```

```python
plt.plot(abs(fft(sig)))
```

```python
dec=pywt.wavedec(sig, wavelet = 'db4', level=4)
```

```python
plt.plot(dec[3])
```
