```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.fft import fft
import pywt
%matplotlib inline
```

```python
# code to generate single frequency signal
def sample_sin():
    sampling_rate = 44100
    duration = 1
    N = sampling_rate*duration
    freq = 20
    omega = 2*np.pi*freq
    t = np.linspace(0,N,250)
    y = np.sin(omega*t)
    plt.plot(t,y)
    plt.show
    return y
```

```python
y = sample_sin()
```

```python
y_dash = fft(y)
```

```python
plt.plot(abs(y_dash))
```
